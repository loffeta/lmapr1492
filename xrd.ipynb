# Diffractogramme de l'oxyde de magnésium

```python
from pymatgen import MPRester #Pour importer le matériau (MgO)
import matplotlib.pyplot as plt 

with MPRester("0GxLMrZ3A1RsWdHOh") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-1009129")
```

```python
#Calcul du modèle de diffraction 
from pymatgen.analysis.diffraction.xrd import XRDCalculator 
Calculator= XRDCalculator('CuKa') # Création du calculateur de diffractométrie de rayons X CuKa
Pattern= Calculator.get_pattern(structure)
```

```python
#Cellule pour l'affichage
print("Voici les 3 premiers pics du diffractogramme pour un rayonnement X incident CuKa:")
Calculator.get_plot(structure,(0, 40))# Création du graphe
plt.show()
print("Les premiers pics apparaissent pour des angles de:", Pattern.x[0:3])
print("L'intensité de l'onde diffractée pour ces angles est de:",Pattern.y[0:3])
print("Les indices de Miller associés sont:")
print("Pour l'angle de", Pattern.x[0],Pattern.hkls[0])
print("Pour l'angle de", Pattern.x[1],Pattern.hkls[1])
print("Pour l'angle de", Pattern.x[2],Pattern.hkls[2])
print("Comme vous l'avez remarqué les indices indices de Miller sont définis par 4 indices.")
print("Ceci est dû à la structure hexagonale de l'oxyde de magnésium.")
print("En effet, c'est ici la notation de Bravais-Miller propre aux structures hexagonales et trigonales qui est utilisée.")
print("Elle a la forme (hkil) avec i = -h-k.")
print("Les distances interplanaires dans ces directions sont (en angstrom):",Pattern.d_hkls[0:3])
```
